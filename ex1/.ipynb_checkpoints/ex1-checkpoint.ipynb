# Part 2: Plotting

```python
# load data
data = np.loadtxt('ex1data1.txt', delimiter=',')
N = data.shape[0]
# notice that both X and y are two-dimensional column vector
X_ori = data[:, 0].reshape(-1, 1)
y     = data[:, 1].reshape(-1, 1)
# plot data
plt.title('Figure 1: Scatter plot of training data')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.plot(X_ori, y, 'rx', label='Training data')
plt.legend(loc='lower right')
plt.show()
```

# Part 3: Cost and Gradient descent
Complete function compute_cost in src/ex1.py

```python
# add a column of ones to X
X = np.hstack((np.ones((N, 1)), X_ori))
print('Testing the cost function ...')
J = compute_cost(X, y, np.array([[0], [0]]))
print('With theta = [0 ; 0]')
print('Cost computed = %f' % J)
print('Expected cost value (approx) 32.07')
J = compute_cost(X, y, np.array([[-1], [2]]))
print('With theta = [-1 ; 2]')
print('Cost computed = %f' % J)
print('Expected cost value (approx) 54.24')
```

```python
theta = np.zeros((2, 1))
iterations = 1500
alpha = 0.01
print('Running Gradient Descent ...')
theta, history_J = gradient_descent(X, y, theta, alpha, iterations)
print('Theta found by gradient descent: [ %.4f  %.4f ]' % (theta[0, 0], theta[1, 0])) 
print('Expected theta values (approx): [ -3.6303  1.1664 ]');
```

```python
# Plot the linear fit
plt.title('Figure 2: Training data with linear regression fit')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
xx = np.array([5, 22.5])
yy = xx * theta[1] + theta[0]
plt.plot(X_ori, y, 'rx', label='Training data')
plt.plot(xx, yy, label='Linear regression')
plt.legend(loc='lower right')
plt.show()
```

# Part 4: Visualizing J

First show how J changes against iterations trained.

```python
plt.title('Figure 3: J of theta against iterations trained')
plt.xlabel('Iteration')
plt.ylabel('J')
plt.plot(history_J)
plt.show()
```

Then show the landscape of J with respect to theta0 and theta1.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Make data.
theta0_list = np.linspace(-10, 10, 100)
theta1_list = np.linspace(-1, 4, 100)
theta0_vals, theta1_vals = np.meshgrid(theta0_list, theta1_list, indexing='ij')
J_vals = np.zeros_like(theta0_vals)
for i, theta0 in enumerate(theta0_list):
    for j, theta1 in enumerate(theta1_list):
        J_vals[i, j] = compute_cost(X, y, np.array([[theta0], [theta1]]))

# Plot the surface.
surf = ax.plot_surface(
    theta0_vals,
    theta1_vals,
    J_vals,
    rcount=100,
    ccount=100,
    cmap=cm.jet,
    antialiased=True
)
ax.set_title('Figure 4: Landscape of cost J')
ax.set_xlabel('theta0')
ax.set_ylabel('theta1')
ax.set_zlabel('J')
ax.invert_xaxis()

# Add a color bar which maps values to colors.
# fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
plt.close(fig)
```

```python
plt.title('Figure 5: Contour of cost J')
plt.xlabel('theta0')
plt.ylabel('theta1')
c = plt.contour(theta0_vals, theta1_vals, J_vals, levels=np.logspace(-2, 3, 20), cmap=cm.jet)
# plt.clabel(c)
plt.plot(theta[0], theta[1], 'rx')
plt.show()
```
