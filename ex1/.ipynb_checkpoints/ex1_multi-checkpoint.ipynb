# Part 0: Setup

```python
import numpy as np
import matplotlib.pyplot as plt
from src.ex1_multi import *
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0)    
```

# Part 1: Feature Normalization
Complete function feature_normalize in src/ex1_multi.py.

```python
print('Loading Data ...')
data = np.loadtxt('ex1data2.txt', delimiter=',')
X_ori = data[:, 0:2]
y = data[:, 2].reshape(-1, 1)
N = X_ori.shape[0]
print('First 10 examples from the dataset:')
print('    x1    x2    y')
for i in range(10):
    print('%-4d%-6.0f%-6.0f%-.0f' % (i + 1, X_ori[i, 0], X_ori[i, 1], y[i, 0]))
print('Normalizing Features ...')
X_gd, mu, sigma = feature_normalize(X_ori)
X_gd = np.hstack((np.ones((N, 1)), X_gd)) # add a column of one to X
```

# Part 2: Gradient Descent
Complete function gradient_descent_multi in src/ex1_multi.py.

```python
print('Running gradient descent ...')
alpha = 0.01
num_iters = 1500
theta = np.zeros((3, 1))
theta, history_J = gradient_descent_multi(X_gd, y, theta, alpha, num_iters)
plt.title('Figure 1: Convergence graph')
plt.xlabel('Number of iterations')
plt.ylabel('Cost J')
plt.plot(history_J)
plt.show()
print('Theta computed from gradient descent: [ %f  %f  %f ]' % (theta[0, 0], theta[1, 0], theta[2, 0]))
```

```python
# Estimate the price of a 1650 sq-ft, 3 br house
Xx = np.array([[1650, 3]])
Xx = np.hstack((np.ones((1, 1)), (Xx - mu) / sigma))
price = (Xx @ theta)[0, 0]
print('Predicted price of a 1650 sq-ft, 3 br house (using gradient descent): %f' % price)
```

# Part 3: Normal Equations

```python
# Use normal equation to solve theta
X_ne = np.hstack((np.ones((N, 1)), X_ori)) # add a column of one to X
theta_ne = normal_equation(X_ne, y)
print('Theta computed from normal equation: [ %f  %f  %f ]' % (theta_ne[0, 0], theta_ne[1, 0], theta_ne[2, 0]))
# Estimate the price of a 1650 sq-ft, 3 br house
Xx = np.array([[1, 1650, 3]])
price = (Xx @ theta_ne)[0, 0]
print('Predicted price of a 1650 sq-ft, 3 br house (using normal equation): %f' % price)
```

# Part 4: Tuning Learing Rate

```python
num_iters = 400
theta = np.zeros((3, 1))
plt.title('Figure 2: Different convergence graph')
plt.xlabel('Number of iterations')
plt.ylabel('Cost J')
for alpha in [0.003, 0.01, 0.03, 0.09, 0.3, 0.9]: # alpha=3 will cause loss to explode
    _, history_J = gradient_descent_multi(X_gd, y, theta, alpha, num_iters)
    plt.plot(history_J, label=('alpha=%f' % alpha))
plt.legend()
plt.show()
```

**Note THAT if use unnormalized X directly in gradient descent, loss J is easier to explode (around alpha=0.03).**
